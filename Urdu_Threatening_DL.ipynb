### LSTM

```python
Embedding layer=10000
dropout=0.5
Dense=10 neurons with relu
LSTM = 100, tanh
dense = 1 with sigmoid 
```

```python

#Deep Learning Network Structure
model_lstm = Sequential()
model_lstm.add(Embedding(10000,100, input_length=X.shape[1]))
model_lstm.add(Dropout(0.5))
#model_conv.add(Bidirectional(GRU(32)))

#odel_conv.add(GRU(units=50, return_sequences=True, input_shape=(X.shape[1],1), activation='relu'))
model_lstm.add(Dense(10,activation='relu'))

model_lstm.add(LSTM(100))
model_lstm.add(Dense(1, activation='sigmoid'))
model_lstm.compile(loss='binary_crossentropy', optimizer='adam',   metrics=['accuracy'])
#Y = data.Reviewer_Score
#Y = pd.get_dummies(dataF['TextBlob Sentiment']).values

X_train, X_valid, Y_train, Y_valid = train_test_split(X,y, test_size = 0.3, random_state =2)
batch_size=128
#Here we train the Network.

model_lstm_pred=model_lstm.fit(X_train, Y_train, batch_size =batch_size, epochs =100, verbose =2,validation_data=(X_valid,Y_valid))
```

```python
model_lstm.summary()
```

```python
pred_lstm=model_lstm.predict(X_valid)
#pred = np.argmax(np.round(pred),axis=1)
pred_lstm = np.round_(pred_lstm)

from sklearn.metrics import confusion_matrix as cm
cm(Y_valid, pred_lstm)


from sklearn.metrics import f1_score

cm1 = confusion_matrix(Y_valid,pred_lstm)
print (cm1) 
tp=cm1[0,0]
fp=cm1[0,1]
fn=cm1[1,0]
tn=cm1[1,1]

pr=tp/(tp+fp)
rc=tp/(tp+fn)
#f1=2*(pr*rc)/(pr+rc)
print("LSTM")
print("Accuracy",metrics.accuracy_score(Y_valid, pred_lstm)*100)
print("Precision", pr*100)
print("RC", rc*100)
print("F1",f1_score(Y_valid, pred_lstm, average='macro')*100)

std=model_lstm_pred.history['accuracy']
import statistics
print(statistics.stdev(std)) 
#predict_prob=model_conv.predict([testa,testb])

#predict_classes=np.argmax(predict_prob,axis=1)
from sklearn.metrics import roc_curve, roc_auc_score
probs=model_lstm.predict(X_valid)
# calculate roc curves
rf_probs = probs
rf_probs = rf_probs[:, 0]
rf_auc = roc_auc_score(Y_valid, rf_probs)
rf_fpr, rf_tpr, threshold = roc_curve(Y_valid, rf_probs)
plt.figure(figsize=(10, 5), dpi=100)
plt.plot([0, 1], [0, 1], linestyle="--", lw=2, color="r", label="Chance", alpha=0.8)
plt.plot(rf_fpr, rf_tpr, marker='.', label='LR (auc = %0.3f)' % rf_auc)
print(rf_auc)
```

```python

#Deep Learning Network Structure
model_conv = Sequential()
model_conv.add(Embedding(10000,100, input_length=X.shape[1]))
model_conv.add(Dropout(0.5))
model_conv.add(Bidirectional(GRU(32)))

#odel_conv.add(GRU(units=50, return_sequences=True, input_shape=(X.shape[1],1), activation='relu'))
model_conv.add(Dense(10,activation='relu'))

#model_conv.add(LSTM(100))
model_conv.add(Dense(1, activation='sigmoid'))
model_conv.compile(loss='binary_crossentropy', optimizer='adam',   metrics=['accuracy'])
#Y = data.Reviewer_Score
#Y = pd.get_dummies(dataF['TextBlob Sentiment']).values

X_train, X_valid, Y_train, Y_valid = train_test_split(X,y, test_size = 0.3, random_state =2)
batch_size=128
#Here we train the Network.

pred1=model_conv.fit(X_train, Y_train, batch_size =batch_size, epochs =100, verbose =2,validation_data=(X_valid,Y_valid))
```

### CNN

```python
def CNN():
        inputs = Input(name='inputs',shape=[X.shape[1]])
        layer = Embedding(10000,100,input_length=X.shape[1])(inputs)
        layer=Conv1D(32, 3, padding='same', activation='relu')(layer)
        layer = MaxPooling1D()(layer)
        layer= Flatten()(layer)
        layer = Dense(128,activation='relu')(layer)
        layer = Dense(1,activation='sigmoid')(layer)
        model_cnn = Model(inputs=inputs,outputs=layer)
        return model_cnn
model_cnn = CNN()
model_cnn.summary()
model_cnn.compile(loss='binary_crossentropy',optimizer=RMSprop(),metrics=['accuracy'])
```

```python
sd=model_cnn.fit(X_train,Y_train,
          epochs=50,
          validation_split=0.1,
          initial_epoch=3)
```
